```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.image as mpimg
```

```python
img = cv2.imread('image5.jpeg',0)
cv2.imshow('Given image',img)
plt.imshow(img, 'gray')
```

```python
ret, thresh = cv2.threshold(img, 70, 155, cv2.THRESH_BINARY)
ret, thresh_inv = cv2.threshold(img, 70, 155, cv2.THRESH_BINARY_INV)
```

```python
title = ['Original', 'Binary', 'Inverted Binary']
image = [img, thresh, thresh_inv]

for i in range(3):
    plt.subplot(2,3,i+1), plt.imshow(image[i],'gray')
    plt.title(title[i])
    plt.xticks([]), plt.yticks([])
```

```python
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
```

```python
large_areas = sorted(contours, key=cv2.contourArea)
```

```python
mask = np.zeros(img.shape,np.uint8)
```

```python
img_contour = cv2.drawContours(mask, [large_areas[-1]], 0, (255,255,255), -1)
```

```python
title = ['Original', 'Binary', 'Inverted Binary', 'Masked']
image = [img, thresh, thresh_inv, img_contour]

for i in range(4):
    plt.subplot(2,3,i+1), plt.imshow(image[i],'gray')
    plt.title(title[i])
    plt.xticks([]), plt.yticks([])
```
