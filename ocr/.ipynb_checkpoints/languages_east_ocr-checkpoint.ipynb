```python
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
from IPython.display import Image
import time
```

# telugu

```python
#----------Method-1---------- "just one word"

start = time.time()
image = 'telugu.png'
reader = easyocr.Reader(['en','te'])
result = reader.readtext(image,paragraph="False")
result

top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
img = cv2.imread(image)
spacer = 45
for detection in result: 
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text = detection[1]
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
    img = cv2.putText(img,text,(15,spacer), font, 0.4,(0,255,0),2,cv2.LINE_AA)
    spacer+=35

plt.figure(figsize=(10,10))
print("\n\n\nDetected text: " + text)
print(type(text))
plt.imshow(img)
plt.show()
out = str(time.time()-start)
print("time taken to detect text: " + out)
```
