```python
#required libraries

import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import colors
from matplotlib.colors import hsv_to_rgb
```

```python
#read the image
#cv2 reads image in BRG format so convert input image into RGB

image = cv2.imread('image3.png')

plt.imshow(image)
plt.show()
```

```python
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
#to know the intensity of colors in given image, a 3D plot should be plot for RGB values

r, g, b = cv2.split(image)
fig = plt.figure()
pixel_colors = image.reshape((np.shape(image)[0]*np.shape(image)[1], 3))
norm = colors.Normalize(vmin=-1.,vmax=1.)
norm.autoscale(pixel_colors)
pixel_colors = norm(pixel_colors).tolist()
axis = fig.add_subplot(1, 1, 1, projection="3d")
axis.scatter(r.flatten(), g.flatten(), b.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Red")
axis.set_ylabel("Green")
axis.set_zlabel("Blue")
plt.show()
```

```python
#the given image is in RGB format but we need to convert it to HSV for better performance

hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

h, s, v = cv2.split(hsv_image)
fig = plt.figure()
axis = fig.add_subplot(1, 1, 1, projection="3d")

axis.scatter(h.flatten(), s.flatten(), v.flatten(), facecolors=pixel_colors, marker=".")
axis.set_xlabel("Hue")
axis.set_ylabel("Saturation")
axis.set_zlabel("Value")
plt.show()
```

```python
#define a lower and upper boundary for color ranges to be detected
#values correspond: HSV (Hue, Saturation, Volume)
#define lower and upper values, then display those for reference

lower_boundary = (0, 44, 15)
upper_boundary = (19, 255, 255)

lb_square = np.full((10, 10, 3), lower_boundary, dtype=np.uint8) / 255.0
ub_square = np.full((10, 10, 3), upper_boundary, dtype=np.uint8) / 255.0

plt.subplot(1, 2, 2)
plt.imshow(hsv_to_rgb(lb_square))
plt.subplot(1, 2, 1)
plt.imshow(hsv_to_rgb(ub_square))
plt.show()
```

```python
#now we have to mask these colors for our required result
#these colors will be kept in the image and rest all would be removed
#save the resultant image as a new image

mask = cv2.inRange(hsv_image, lower_boundary, upper_boundary)
result = cv2.bitwise_and(image, image, mask=mask)
plt.subplot(1, 2, 1)
plt.imshow(mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()
result=cv2.cvtColor(result, cv2.COLOR_BGR2RGB)
cv2.imwrite('result.jpeg', result)
print(mask.shape)
```

```python
#every white pixel in the image corresponds to the color given in the lower and upper boundaries
#shape function gives first 2 values are size of image, last value is dimensions
#divide the mask image size with real image size, we get white pixels percentage, which is the value of 
#brown peel required

final_image = cv2.imread('result.jpeg')
final_image=cv2.cvtColor(final_image, cv2.COLOR_BGR2RGB)
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.subplot(1, 2, 2)
plt.imshow(final_image)
plt.show()

height, width = mask.shape[:2]
#get total pixels of image
pixels = height * width
white_pixels = cv2.countNonZero(mask)
percent_white = (white_pixels/pixels) * 100
percent_white = round(percent_white,2)
print('potato peeled percentage: ', 100-percent_white) #percentage of potato peeled
```
